```python
# --- Import required libraries ---
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# --- Generate training data ---
np.random.seed(42)
angles = np.random.uniform(10, 80, 500)        # degrees
velocities = np.random.uniform(5, 60, 500)     # m/s
g = 9.81

# Calculate theoretical data for training
max_heights = (velocities**2 * (np.sin(np.radians(angles)))**2) / (2 * g)
distances = (velocities**2 * np.sin(np.radians(2*angles))) / g

# Create dataset
data = pd.DataFrame({
    "angle": angles,
    "velocity": velocities,
    "max_height": max_heights,
    "distance": distances
})

# --- Train models ---
X = data[["angle", "velocity"]]
y_height = data["max_height"]
y_distance = data["distance"]

X_train, X_test, y_train_h, y_test_h = train_test_split(X, y_height, test_size=0.2, random_state=42)
X_train, X_test, y_train_d, y_test_d = train_test_split(X, y_distance, test_size=0.2, random_state=42)

model_height = LinearRegression().fit(X_train, y_train_h)
model_distance = LinearRegression().fit(X_train, y_train_d)

print("âœ… Models trained successfully!\n")

# --- Get user input ---
angle = float(input("Enter the angle of projection (degrees): "))
velocity = float(input("Enter the initial velocity (m/s): "))

# --- ML predictions ---
pred_height = model_height.predict([[angle, velocity]])[0]
pred_distance = model_distance.predict([[angle, velocity]])[0]

# --- Theoretical results (for comparison) ---
true_height = (velocity**2 * (np.sin(np.radians(angle)))**2) / (2 * g)
true_distance = (velocity**2 * np.sin(np.radians(2*angle))) / g

# --- Display results ---
print("\nðŸŽ¯ Predicted (ML) Results:")
print(f"Maximum Height: {pred_height:.2f} meters")
print(f"Horizontal Distance: {pred_distance:.2f} meters")

print("\nðŸ“˜ Theoretical (Physics) Results:")
print(f"Maximum Height: {true_height:.2f} meters")
print(f"Horizontal Distance: {true_distance:.2f} meters")

# --- Plot projectile trajectory ---
# We'll use the theoretical motion equation for the trajectory shape
t_flight = 2 * velocity * np.sin(np.radians(angle)) / g
t = np.linspace(0, t_flight, num=100)
x = velocity * np.cos(np.radians(angle)) * t
y = velocity * np.sin(np.radians(angle)) * t - 0.5 * g * t**2

plt.figure(figsize=(9,5))
plt.plot(x, y, label='Projectile Path', color='dodgerblue', linewidth=2)
plt.scatter([pred_distance/2], [pred_height], color='red', label='Predicted Peak (ML)')
plt.title(f'Projectile Motion (Angle={angle}Â°, Velocity={velocity} m/s)')
plt.xlabel('Horizontal Distance (m)')
plt.ylabel('Height (m)')
plt.legend()
plt.grid(True)
plt.show()
```
